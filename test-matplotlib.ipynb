```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 5, 0.01) #create x [0..5] in 0.1 increments
y = np.sin(x)            # y = sin(x)
plt.plot(x,y, color='purple')            #make a plot
plt.xlabel('x')          #label x axis
plt.ylabel('sin(x)')     #label y axis
plt.savefig('sinx.png' ,bbox_inches="tight" ,dpi=600)               #save the figure
```
