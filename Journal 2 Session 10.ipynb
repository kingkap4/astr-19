```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.standard_exponential(1000)

width = 0.01
histmin = np.floor(min(x))
histmax = np.ceil(1)+width

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=bins,alpha=0.5,edgecolor="black")
plt.ylabel("y")
plt.xlabel("x")

plt.savefig("journal9.pdf" ,bbox_inches="tight" ,facecolor="white")
```
